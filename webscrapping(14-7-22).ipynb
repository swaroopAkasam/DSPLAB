```python
import requests
from bs4 import BeautifulSoup
x=requests.get("https://www.geeksforgeeks.org/python-web-scraping-tutorial/")
soup=BeautifulSoup(x.content,"html.parser")
x=[]
for i in soup.find_all("img"):
    s=i.get("src")
    a=i.get("alt")
    x.append({"src":s,"alt":a})
for j in x:
    print(j)

```
